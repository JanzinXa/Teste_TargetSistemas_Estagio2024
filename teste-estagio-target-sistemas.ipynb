```python
def pertence_sequencia_fibonacci(numero):
    # Inicializa os dois primeiros números da sequência de Fibonacci
    a, b = 0, 1
    while a <= numero:
        # Verifica se o número informado é igual ao número atual da sequência
        if a == numero:
            return True
        # Calcula os próximos números da sequência de Fibonacci
        a, b = b, a + b
    # Se o número não for encontrado na sequência, retorna False
    return False

while True:
    # Solicita ao usuário um número para verificar se pertence à sequência de Fibonacci
    entrada = input("Digite um número para verificar se pertence à sequência de Fibonacci (ou digite 'x' para sair): ")

    # Verifica se o usuário quer sair
    if entrada.lower() == 'x':
        print("Encerrando o programa.")
        break

    # Verifica se a entrada é um número ou não
    try:
        numero_informado = int(entrada)
    except ValueError:
        print("Entrada inválida. Por favor, digite um número válido ou 'x' para sair.")
        continue

    # Verifica se o número pertence à sequência de Fibonacci e exibe a mensagem correspondente
    if pertence_sequencia_fibonacci(numero_informado):
        print(f"O número {numero_informado} pertence à sequência de Fibonacci.")
    else:
        print(f"O número {numero_informado} NÃO pertence à sequência de Fibonacci.")
```

```python
# Função para inverter uma string
def inverter_string(string):
    # Inicializa uma string vazia para armazenar o resultado
    string_invertida = ""
    
    # Itera sobre os caracteres da string de trás para frente
    for i in range(len(string) - 1, -1, -1):
        # Adiciona o caractere atual ao resultado
        string_invertida += string[i]
    
    # Retorna a string invertida
    return string_invertida

# Loop principal
while True:
    # Pergunta ao usuário o que deseja inverter ou se deseja sair
    opcao = input("Digite a string que deseja inverter ou 'x' para sair: ")
    
    # Verifica se o usuário deseja sair
    if opcao.lower() == 'x':
        print("Saindo do programa...")
        break
    
    # Inverte a string fornecida pelo usuário e imprime o resultado
    string_invertida = inverter_string(opcao)
    print("String original:", opcao)
    print("String invertida:", string_invertida)
```
